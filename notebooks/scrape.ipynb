## Scrape collected URLs

**Fields of interest**: Pub date, reporter name, reporter title, article hed, article dek, article text

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

from tqdm import tqdm
```

```python
df = pd.read_csv('../data/COMPOSITE_29days.csv')
```

```python
results = []
for i in tqdm(df.urls):
    r = requests.get(i)
    soup = BeautifulSoup(r.content)
    try:
        byline = soup.find('span', {'class': 'news-byline-full__name'}).get_text()
    except:
        byline = None
    try:
        title = soup.find('span', {'class': 'news-byline-full__role'}).get_text()
    except:
        title = None
    try:
        pub_date = soup.find('p', {'class': 'news-article-header__timestamps-posted'}).get_text()
    except:
        pub_date = None
    try:
        hed = soup.find('h1').get_text()
    except:
        hed = None
    try:
        dek = soup.find('p', {'class': 'news-article-header__dek'}).get_text()
    except:
        dek = None
    try:
        body = '\n'.join([i.get_text() for i in soup.find('div', {'class': 'js-article-wrapper'}).find_all(['h2', 'p'])])
    except:
        body = None
    res = {'byline': byline,
           'title': title,
           'pub_date': pub_date,
           'hed': hed,
           'dek': dek,
           'body': body}
    results.append(res)
```

```python
results_df = pd.DataFrame(results)
```

```python
results_df.to_csv('../data/scraped_data.csv', index=False)
```

```python
results_df = results_df[~results_df['pub_date'].isna()]
```

```python
results_df['parsed_date'] = pd.to_datetime(results_df['pub_date'].apply(lambda x: x.split('Posted on ')[1].split(', at')[0]))
```
